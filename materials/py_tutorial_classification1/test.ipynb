```python
import pandas as pd
import altair as alt
```

```python
can_lang = pd.read_csv("D:/Data/Git/dsci-100-student-python/materials/py_tutorial_classification1/data/can_lang.csv")
```

```python
can_lang.head()
```

```python
# Select only the rows where the category is "Aboriginal languages"
can_lang[can_lang['category'] == "Aboriginal languages"]
# select only the language and mother tongue columns "
can_lang[["language", "mother_tongue"]]
```

```python
# use .loc to filter rows and columns at the same time
aboriginal_lang = can_lang.loc[can_lang['category'] == "Aboriginal languages", ["language", "mother_tongue"]]

arranged_lang  = aboriginal_lang.sort_values("mother_tongue", ascending=False)
ten_lang = arranged_lang.head(10).copy()
ten_lang
```

```python
canadian_population = can_lang["mother_tongue"].sum()
canadian_population
ten_lang["percent_mother_tongue"] = ten_lang["mother_tongue"] / canadian_population * 100
ten_lang
```

```python
alt.Chart(ten_lang).mark_bar().encode(
    x=alt.X("language", title="Top 10 Aboriginal languages in Canada"),
    y=alt.Y("percent_mother_tongue", title="Percentage of Mother Tongue Speakers")
)
```

```python
barplot_mother_tongue_axis = alt.Chart(ten_lang).mark_bar().encode(
    x=alt.X("mother_tongue").title("Mother Tongue (Number of Canadian Residents)"),
    y=alt.Y("language").sort("-x").title("Language")
)
barplot_mother_tongue_axis
```
